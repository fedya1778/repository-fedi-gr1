СОЗДАНИЕ МАССИВОВ

```python
import numpy as np

# Создание массива из обычного списка Python
list_data = [1, 2, 3, 4, 5]
array_from_list = np.array(list_data)
print("Массив из списка:", array_from_list)

# Создание двумерного массива из вложенных списков
matrix_data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix = np.array(matrix_data)
print("Двумерный массив:")
print(matrix)

# Создание массива из нулей
zeros_array = np.zeros(5)
print("Массив из нулей:", zeros_array)

# Создание массива из единиц
ones_array = np.ones((2, 3))  # 2 строки, 3 столбца
print("Массив 2x3 из единиц:")
print(ones_array)

# Создание массива с последовательностью чисел
range_array = np.arange(0, 10, 2)  # от 0 до 10 с шагом 2
print("Массив с arange:", range_array)
```

АТРИБУТЫ МАССИВОВ

```python


print("Форма матрицы:", matrix.shape)    # Размерность массива
print("Тип данных матрицы:", matrix.dtype) # Тип элементов
print("Количество измерений:", matrix.ndim) # Число осей
print("Общее количество элементов:", matrix.size) # Всего элементов

```

ИНДЕКСИРОВАНИЕ

```python
# Получение отдельного элемента (индексация начинается с 0)
print("Элемент [1,2]:", matrix[1, 2])  # Вторая строка, третий столбец

# Срезы - получение подмассива
print("Первые две строки:")
print(matrix[:2])  # Все столбцы, первые две строки

print("Последний столбец:")
print(matrix[:, -1])  # Все строки, последний столбец

print("Подмассив 2x2 из центра:")
print(matrix[1:3, 0:2])  # Строки 1-2, столбцы 0-1
```

ВЕКТОРНЫЕ ОПЕРАЦИИ

```python
# Создадим два массива для операций
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

# Поэлементное сложение
sum_arrays = a + b
print("Сложение a + b:", sum_arrays)

# Поэлементное умножение
mult_arrays = a * b
print("Умножение a * b:", mult_arrays)

# Математические функции применяются к каждому элементу
squared = np.square(a)
print("Квадраты элементов a:", squared)
```

СТАТИСТИЧЕСКИЕ ФУНКЦИИ

```python
data = np.array([2, 4, 6, 8, 10, 12])

print("Исходный массив:", data)
print("Среднее значение:", np.mean(data))
print("Стандартное отклонение:", np.std(data))
print("Максимальное значение:", np.max(data))
print("Минимальное значение:", np.min(data))
print("Сумма всех элементов:", np.sum(data))

# Статистика по осям для матрицы
print("Исходная матрица:")
print(matrix)
print("Среднее по столбцам:", np.mean(matrix, axis=0))  # По вертикали
print("Среднее по строкам:", np.mean(matrix, axis=1))   # По горизонтали
```
